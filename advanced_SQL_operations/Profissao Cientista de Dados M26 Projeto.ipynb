# Projeto de Dashboard de E-commerce

## 1 Objetivo

Este projeto tem como objetivo consolidar dados de transações e clientes de um e-commerce por meio de SQL, exportar a base tratada em CSV e criar um dashboard no Power BI para análise das principais métricas de vendas e perfil dos clientes.

## 2 Carregamento das bases

```python
import sqlite3
import pandas as pd
```

```python
df_transacoes = pd.read_csv("TB_TRANSACOES_PROJETO_ECOMM.csv", delimiter=';')
df_clientes = pd.read_csv("TB_CLIENTES_PROJETO_ECOMM.csv", delimiter=';')
```

```python
conn = sqlite3.connect('projeto.db')
# Carregar o DataFrame no banco de dados SQLite - criando tb_transacoes e tb_clientes
df_transacoes.to_sql('tb_transacoes', conn, index=False, if_exists='replace')
df_clientes.to_sql('tb_clientes', conn, index=False, if_exists='replace')
```

```python
# Função para executar consultas SQL e retornar o resultado como um DataFrame
def run_query(query):
    return pd.read_sql_query(query, conn)
```

## 3. Visualização inicial dos dados

```python
query = """
SELECT *
FROM tb_transacoes
"""
result_df = run_query(query)
print(result_df)
```

```python
query = """
SELECT *
FROM tb_clientes
"""
result_df = run_query(query)
print(result_df)
```

## 4. Consolidação dos dados com SQL

```python
query = """
SELECT 
    A.id_client,
    A.Category,
    CAST(REPLACE(A.Price, ',', '.') AS REAL) AS Price,
    A."Card Type",
    B.state_name,
    B.First_name,
    B.Gender,
    B.Job_Title
FROM tb_transacoes AS A LEFT JOIN tb_clientes AS B ON A.id_client = B.Id_client
"""
result_df = run_query(query)
print(result_df)
```


## 5. Justificativa do JOIN escolhido

Utilizei o LEFT JOIN tendo a tabela de transações como tabela principal, pois o objetivo do projeto é analisar as vendas realizadas. Dessa forma, todas as transações são preservadas no conjunto final, mesmo quando algum cliente não possui correspondência na tabela de dados pessoais.

A tabela de clientes complementa a análise com informações como estado, gênero e cargo, enquanto a tabela de transações concentra os dados principais para calcular métricas como total de vendas, número de transações, ticket médio e categorias mais vendidas.

## 6. Exportação da base consolidada

```python
result_df.to_csv('dados_ecommerce_final.csv', index=False)
```

```python
df_final = pd.read_csv('dados_ecommerce_final.csv')
df_final.head()
```

Após a realização do LEFT JOIN entre as tabelas de transações e clientes, os dados consolidados foram exportados para o arquivo dados_ecommerce_final.csv. Esse arquivo será utilizado no Power BI para a criação do dashboard interativo com métricas como total de vendas, número de transações, clientes únicos, ticket médio, vendas por categoria, estado, gênero e tipo de cartão.
